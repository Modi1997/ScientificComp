## **Exercise 4** <a id="Exercise-4"></a>

Plot the function $y = cos(x)$ over the domain $[0, 2π)$ in intervals of ($5 \times 10^{-2}$).


### Solution
To plot the function $y = cos(x)$ over the domain $[0, 2π)$ in intervals of ($5 \times 10^{-2}$) the following steps are taken:
1. **Define the Domain**: Variables have been created (to keep the code clean and provide easier readability) for the domain such as `start`, `end` and `step`. The values assigned to the variables are based on the given data of the exercise.
2. **Generate x values**: The variables created from the above step have been passed as parameters to the `arrange()` function of NumPy to generate the array with the requested domain and interval.
3. **Generate y values**: The corresponding `y` values are computed by applying the cosine function to each `x` value using the `cos()` function of NumPy.

The goal of this exercise is the use of mathematical notation in the labels to provide a clear representation of the function and domain. Below is an explanation of the key LaTeX-style components that have been used:
1. `$` **for mathematical notation**: The dollar sign is used to indicate that the text inside it should be rendered as mathematical notation.
2. `r` **for raw strings**: The r before the string (e.g., r'...') tells Python that the string is a raw string. This prevents Python from interpreting backslashes \ as escape characters, which is necessary when using LaTeX-style commands in strings.
3. `\command` **for LaTeX commands**: The use of backslash \ followed by a command to call LaTeX functions. For instance, in the title I called \times which is used to display the multiplication symbol (×) in mathematical notation.
4. `{}` **for grouping in LaTeX**: Curly braces {} are used to group elements in LaTeX. In this case, they are used to ensure that the exponent is formatted correctly. If the power is not negative as per the y-axis, then the braces can still be used but are not necessary.
5. `[ )` **for inclusive and exclusive**: Square brackets [ are used to indicate an inclusive boundary, and round parentheses ) to indicate an exclusive boundary. In this case, [0, 200) means that the domain includes 0 but does not include 200.

By using these elements, it is possible to format the labels in a mathematically accurate and visually clear way, making the plot more informative.

```python
# Start variable of the domain 
start = 0

# End variable of the domain
end = 200

# Interval (step) variable of the domain
step = 1 * (10**-2)

# Generate values for 'x' based on 'start', 'end' and 'step'
x = np.arange(start, end, step)

# Generate values for y=100x-x^2
y = (100 * x) - (x**2)

# Plot the function
plt.plot(x, y, label='$y = 100x - x^2$')

# Label the graph
plt.title(r'Plot of $y = 100x - x^2$ over $[0,200)$ in intervals of $1 \times 10^{-2}$')
plt.xlabel('$x \in [0, 200)$')
plt.ylabel('$y = 100x - x^2$')
plt.legend()

# Display the plot
plt.show()
```

Create a variable representing the matrix $$\begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}$$

### Solution
For this task, we first need to assign the given matrix to a variable. The variable is named matrix, and the array is created using NumPy's `array()` function. The brackets within the function define the dimensions of the array. Since we are creating a 2D array, we will use one set of outer brackets to represent the entire matrix. Inside these outer brackets, we will add two inner sets of brackets, one for each row. Each row is separated by a comma, and the values within each row are also separated by commas.

Compute $Y \times \begin{pmatrix} 1 \\ 2 \end{pmatrix}$, where $Y$ is the result from the previous exercise.


### Solution
For this exercise the use of the already existing variable `Y` will be used (2x2) and multiplied with a new (2x1) array to get the result. NumPy has a function called `dot()` which computes the product of 2 arrays for you. In this case we will pass as **parameters** to this function the 'Y' and 'array' variables (arrays). 

The result is saved in a new variable called `result` which is later on printed. In this case the the expected result would be $[[0.0], [0.5]]$ however the result displayed is $[[2.22044605e-16], [5.00000000e-01]]$. The actual value of this numbers is $2.22044605×10^{-16}$ which is actually a very small number which can be considered as $0$ and $5.00000000 \times 10^{-1}$ which in fact is $0.5$. 

In terms of **why is this happening** is because NumPy uses floating-point arithmetic to perform numerical operations, which can lead to small precision errors due to the way computers store numbers in binary form. These errors, such as returning very small values like $2.22044605×10^{-16}$ instead of exactly $0$, occur because not all numbers can be represented exactly in binary. NumPy returns the most precise result possible to ensure accuracy and provides a function named `round()` which allows users to specify both the variable and the desired number of decimal places for rounding, providing flexibility in how results are displayed.

Below, both the original array with the very small value and the array rounded to one decimal place are printed for comparison.

```python
# Define the array
array = np.array([[1], [2]])

# Compute the matrix multiplication of Y and new array
result = np.dot(Y, array)

# Print the result of the matrix multiplication
print("The result of Y * [1, 2] is:")
print(result)

# Result rounded to 1 decimal
print("\n")
print("The result of Y * [1, 2] rounded to 1 decimal is:")
print(np.round(result, decimals=1))
```

The resulting plot is a line chart that visualizes the cumulative sum of the `new_sequence`. It begins either with -1 or 1 (depending on the first element of the array) and then adds subsequent values (-1 or +1) to create a cumulative plot through the $100_{th}$ random element. Next, the `np.where()` function replaces all occurrences of 0 with -1, resulting in a sequence where the values are either -1 or 1. Also, as before, both outcomes have equal probability (50% each).

Then the `np.cumsum()` function is then applied to `realizations_minus_plus` with `axis=1`, which calculates the cumulative sum for each step on each realization. ## **Exercise 18** <a id="Exercise-18"></a>

Write a Python function that implements the quadratic formula for finding the roots of a quadratic function, in such a way that it does not suffer from at least one of the two potential sources of catastrophic cancellation.

*Hint: Consult David Goldberg's paper for this, which is mentioned in the slides and available in the Reading Material section on KEATS. Study in particular the section on catastrophic cancellation.*


### Solution
The function `quadratic_formula` is defined to take three numerical inputs `a`, `b`, and `c`, representing the coefficients of a quadratic equation $ax^2 + bx + c = 0$. Within the function, the following steps are taken to calculate the 2 roots and avoid catastrophic cancellation:
1. **Discriminant Calculation**: The first step is computing the discriminant, $\Delta = b^2 - 4ac$. A condition checks if the discriminant is negative, and if so it means that no real roots exist in that case the function returns `(None, None)`.
2. **Handling Catastropic Cancellation**: If the discriminant is non-negative then its square root is computed by using the NumPy's `sqrt()` function and passing the discriminant as an argument. As per David Golberg's paper a catastrophic cancellation occurs when ***[1]***, <a id="Reference-1"></a> *"The quantities b2 and 4ac are subject to rounding errors since they are the results of floating-point multiplications."*. The first to handle the catastrophic cancellation is to check whether `b` is a positive or negative number.
    * If `b` is **positive** &rarr; Root 1 is computed by substracting the square root of the discriminant from `-b` to reduce the cancellation errors. Root 2, an alternative formula is used to avoid precision loss by calculating $\frac{2c}{\sqrt{-b+\Delta}}$.
    * If `b` is **negative** &rarr; To handle the case where `b` is negative, the function adds the square root of the discriminant to `-b` for Root 1 to minimize the cancellation. Similarly, Root 2 is computed using the alternative form $\frac{2c}{\sqrt{-b+\Delta}}$ to avoid precision errors.
    
In conclusion, the `quadratic_formula` ensures numerically stable results by selecting the appropriate operation based on `b` (negative or positive), which minimizes the risk of rounding errors during floating-point operations that can occur in the standard quadratic formula.

Finally, two test cases are provided to illustrate how the function works, displaying the calculated roots of two different quadratic equations. Please note that the values for the first case are taken from David Golberg's paper *'What Every Computer Scientist Should Know About Floating-Point Arithmetic'*.

```python
# Definition of the quadratic formula function that avoids catastrophic cancellation
def quadratic_formula(a, b, c):
    """
    Solves the quadratic equation ax^2 + bx + c = 0 using a numerically stable approach to avoid catastrophic cancellation.
    If the discriminant is negative, the function returns (None, None) to indicate no real solutions.

    Parameters:
        a (float): The coefficient of the quadratic term (x^2)
        b (float): The coefficient of the linear term (x)
        c (float): Constant term

    Returns:
        tuple: Two roots of the quadratic equation (root1, root2)
    """
    
    # Compute the discriminant
    discriminant = b**2 - 4*a*c
    
    if (discriminant < 0):
        # No real roots exist
        return None, None
    
    # Square root of discriminant
    sqrt_discriminant = np.sqrt(discriminant)
    
    # Use a numerically stable form of the quadratic formula
    if b >= 0:
        # If b positive then substruct to minimize cancellation
        root1 = (-b - sqrt_discriminant) / (2*a)
        # Use an alternative form for the second root to avoid precision loss
        root2 = (2*c) / (-b - sqrt_discriminant)
    else:
        # If b is negative then add to minimize cancellation
        root1 = (-b + sqrt_discriminant) / (2*a)
        # Use an alternative form for the second root to avoide precision loss
        root2 = (2*c) / (-b + sqrt_discriminant)
    
    return root1, root2


# Sample 1 to test the function
# The a,b,c values of this test have been selected based on the example from David Goldber's paper about catastrophic cancellation
a, b, c = 1.22, -3.34, 2.28
roots = quadratic_formula(a, b, c)
print(f"The roots of the quadratic equation for a={a}, b={b} and c={c} are: {roots}")

# Sample 2 to test the function
a, b, c = 1, -4, 4.000001
roots = quadratic_formula(a, b, c)
print(f"The roots of the quadratic equation for a={a}, b={b} and c={c} are: {roots}")
```
